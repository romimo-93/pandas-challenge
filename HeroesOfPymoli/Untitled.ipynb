```python
# Dependencies and Setup
import pandas as pd
# File to Load 
csv_file = "../Resources/02-Homework_04-Pandas_Instructions_HeroesOfPymoli_Resources_purchase_data.csv"
# Read Purchasing File and store into Pandas data frame
pymoli_data = pd.read_csv(csv_file)

pymoli_data.head()
```

```python
#Player Count
total_players = len(pymoli_data["SN"].unique())
```

```python
# total_players_df = [{"Player Count": "total_players"}]

player_count_df = pd.DataFrame({"Total Players": [total_players]})
player_demographics = pymoli_data.loc[:,["Gender","SN","Age"]].drop_duplicates()
player_count_df
```

```python
#Purchasing Analysis total
    #Number of Unique Items
unique_items = len(pymoli_data["Item Name"].unique())
    # Average Purchase Price
average_price = pymoli_data["Price"].mean()
average_price = f"${average_price:.2f}"
    # Total Number of Purchases
total_purchases = pymoli_data["Item Name"].count()
    # Total Revenue
total_revenue = pymoli_data["Price"].sum()
total_revenue = f"${total_revenue:.2f}"

summary_df = pd.DataFrame({"Number of Unique Items":[unique_items], "Average Price":[average_price],"Number of Purchases":[total_purchases], "Total Revenue":[total_revenue]})

summary_df
```

```python
#Gender Demographics
    # Percentage and Count of Male Players
unique_male = pymoli_data.loc[pymoli_data["Gender"] =="Male", ["SN"]]
unique_male_SN = len(unique_male["SN"].unique())
unique_male_percentage = (unique_male_SN)/(total_players)*100
unique_male_percentage = f"{unique_male_percentage:.2f}%"     
    # Percentage and Count of Female Players
unique_female = pymoli_data.loc[pymoli_data["Gender"] =="Female", ["SN"]]
unique_female_SN = len(unique_female["SN"].unique())
unique_female_percentage = (unique_female_SN)/(total_players)*100
unique_female_percentage = f"{unique_female_percentage:.2f}%"

    # Percentage and Count of Other / Non-Disclosed
unique_other = pymoli_data.loc[pymoli_data["Gender"] =="Other / Non-Disclosed", ["SN"]]
unique_other_SN = len(unique_other["SN"].unique())
unique_other_percentage = (unique_other_SN)/(total_players)*100
unique_other_percentage = f"{unique_other_percentage:.2f}%"

gender_summary = pd.DataFrame({
    "Total Count":(unique_male_SN,unique_female_SN,unique_other_SN),
    "Percentage of Players":(unique_male_percentage,unique_female_percentage,unique_other_percentage),
    "Gender":("Male","Female","Other")
})

gender_summary.set_index("Gender",inplace=True)

gender_summary
```

```python
# Purchasing Analysis (Gender)
    # The below each broken by gender
# Purchase Count
male_purchase = pymoli_data.loc[pymoli_data["Gender"] =="Male", ["Item Name"]]
male_purchase_count = male_purchase["Item Name"].count()
# Average Purchase Price
male_price = pymoli_data.loc[pymoli_data["Gender"] =="Male", ["Price"]]
male_avg_price = male_price["Price"].mean()
male_avg_price = f"${male_avg_price:.2f}"
# Total Purchase Value
male_total = pymoli_data.loc[pymoli_data["Gender"] =="Male", ["Price"]]
male_purchase_total = male_total["Price"].sum()
# Average Purchase Total per Person by Gender 
male_average_per_person = (male_purchase_total)/(unique_male_SN)

male_purchase_total = f"${male_purchase_total:.2f}"
male_average_per_person = f"${male_average_per_person:.2f}"

# Purchase Count
female_purchase = pymoli_data.loc[pymoli_data["Gender"] =="Female", ["Item Name"]]
female_purchase_count = female_purchase["Item Name"].count()
# Average Purchase Price
female_price = pymoli_data.loc[pymoli_data["Gender"] =="Female", ["Price"]]["Price"].mean()
# female_avg_price = female_price["Price"].mean()
female_avg_price = f"${female_price:.2f}"
# Total Purchase Value
female_total = pymoli_data.loc[pymoli_data["Gender"] =="Female", ["Price"]]
female_purchase_total = female_total["Price"].sum() 
# Average Purchase Total per Person by Gender
female_average_per_person = (female_purchase_total)/(unique_female_SN)

female_purchase_total = f"${female_purchase_total:.2f}"
female_average_per_person = f"${female_average_per_person:.2f}"

# Purchase Count
other_purchase = pymoli_data.loc[pymoli_data["Gender"] =="Other / Non-Disclosed", ["Item Name"]]
other_purchase_count = other_purchase["Item Name"].count()
# Average Purchase Price
other_price = pymoli_data.loc[pymoli_data["Gender"] =="Other / Non-Disclosed", ["Price"]]
other_avg_price = other_price["Price"].mean()
other_avg_price = f"${other_avg_price:.2f}"
# Total Purchase Value
other_total = pymoli_data.loc[pymoli_data["Gender"] =="Other / Non-Disclosed", ["Price"]]
other_purchase_total = other_total["Price"].sum()
# Average Purchase Total per Person by Gender
other_average_per_person = (other_purchase_total)/(unique_other_SN)

other_purchase_total = f"${other_purchase_total:.2f}"
other_average_per_person = f"${other_average_per_person:.2f}"



gender_purchase_analysis = pd.DataFrame({
    "Purchase Count":(male_purchase_count,female_purchase_count,other_purchase_count),
    "Average Purchase Price":(male_avg_price,female_avg_price,other_avg_price),
    "Total Purchase Value":(male_purchase_total,female_purchase_total,other_purchase_total),
    "Avg Total Purchase per Person":(male_average_per_person,female_average_per_person,other_average_per_person),
    "Gender":("Male","Female","Other")
})

gender_purchase_analysis.set_index("Gender",inplace=True)

gender_purchase_analysis
```

```python
# Age Demographics
# The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
age_bins = [0,9,14,19,24,29,34,39,100]
age_labels = ["Less Than 10","10 to 14","15 to 19","20 to 24","25 to 29","30 to 34","35 to 39","Greater than 40"]

#value count off of age range
# Purchase Count
pd.cut(player_demographics["Age"],age_bins, labels=age_labels,include_lowest=True)

pymoli_data["Age Group"] = pd.cut(player_demographics["Age"],age_bins, labels=age_labels)
age_totals = pymoli_data["Age Group"].value_counts()

# Average Purchase Price
age_averages = (age_totals)/(total_players)*100

age_analysis = pd.DataFrame({
    "Total Count":(age_totals),
    "Percentage of Players":(age_averages)
})

age_analysis["Percentage of Players"] = age_analysis["Percentage of Players"].map("{:,.2f}%".format)    

age_analysis = age_analysis.sort_index()

age_analysis
     
        

   
```

```python
# Purchasing Analysis (Age)
    # Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
age_bins = [0,9,14,19,24,29,34,39,100]
age_labels = ["Less Than 10","10 to 14","15 to 19","20 to 24","25 to 29","30 to 34","35 to 39","Greater than 40"]

#purchase count
pymoli_data["Age Group"] = pd.cut(pymoli_data["Age"],age_bins, labels=age_labels)
purchase_age_totals = pymoli_data["Age Group"].value_counts()

#average purchase price
grouped_age_group = pymoli_data.groupby(["Age Group"])
average_purchase_price = grouped_age_group["Price"].mean()

#total purchase value
total_purchase_value = grouped_age_group["Price"].sum()

#Average Purchase Total per Person by Age Group
age_average_per_person = (total_purchase_value)/(age_totals)

purchasing_analysis = pd.DataFrame({
    "Purchase Count":(purchase_age_totals),
    "Average Purchase Price":(average_purchase_price),
    "Total Purchase Value":(total_purchase_value),
    "Avg Total Purchase per Person":(age_average_per_person)
})

purchasing_analysis["Total Purchase Value"] = purchasing_analysis["Total Purchase Value"].map("${:,.2f}".format) 
purchasing_analysis["Average Purchase Price"] = purchasing_analysis["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis["Avg Total Purchase per Person"] = purchasing_analysis["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchasing_analysis


#     Create a summary data frame to hold the results
    # Optional: give the displayed data cleaner formatting
    # Display the summary data frame
```

```python
# Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
        # SN
groupby_SN = pymoli_data.groupby(["SN"])

sum_purchase_value = groupby_SN["Price"].sum()

sum_purchase_value_df = pd.DataFrame(sum_purchase_value)


top_purchase_value = sum_purchase_value_df.sort_values("Price",ascending=False)

top_purchase_df = top_purchase_value.head(5)


# # Purchase Count
sum_purchase_count = groupby_SN["Purchase ID"].count()

sum_purchase_count_df = pd.DataFrame(sum_purchase_count)

merge_df = pd.merge(top_purchase_df,sum_purchase_count_df, on="SN")



# Average Purchase Price
average_purchase = groupby_SN["Price"].mean()

average_purchase_df = pd.DataFrame(average_purchase)


final_merge = pd.merge(merge_df,average_purchase_df, on="SN")

final_top_5_analysis = final_merge.rename(columns={"Price_x":"Total Purchase Value","Purchase ID":"Purchase Count","Price_y":"Average Purchase Price"})


final_top_5_analysis
```

```python
# Most Popular Items
    # Identify the 5 most popular items by purchase count, then list (in a table):
groupby_item = pymoli_data.groupby(["Item Name"])
sum_item = groupby_item["Price"].sum()
sum_item_df = pd.DataFrame(sum_item) 
sum_item_df.head()

top_item_value = sum_item_df.sort_values("Price",ascending=False)                                  
top_item_value = top_item_value.head(5) 


sum_item_count = groupby_item["Purchase ID"].count()
sum_item_count_df = pd.DataFrame(sum_item_count)
merge_item_df = pd.merge(top_item_value,sum_item_count_df, on="Item Name")


item_price = groupby_item["Price"].mean()
item_price_df = pd.DataFrame(item_price)


final_item_merge = pd.merge(merge_item_df,item_price_df, on="Item Name")
final_item_analysis = final_item_merge.rename(columns={"Price_x":"Total Purchase Value","Purchase ID":"Purchase Count","Price_y":"Item Price"})
final_item_analysis
        # Item ID
        # Item Name
        # Purchase Count
        # Item Price
        # Total Purchase Value
```
